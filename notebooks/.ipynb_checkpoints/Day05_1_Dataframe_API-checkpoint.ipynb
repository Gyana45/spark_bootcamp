# string functions

```python
# from pyspark.sql.functions import col, upper, lower, initcap, length, trim,substring, regexp_replace, concat, lit, split

from pyspark.sql.functions import *
data = [
(1, "ankit bansal     ", "ankit.bansal@gmail.com"),
(2, "  rohit sharma ", "rohit@gmail.com"),
(3, " virat kohli", "virat@bcci.in")
]
schema = ["id", "name", "email"]
df = spark.createDataFrame(data, schema)
df_string = df.select(
"id",
"name",
"email",
trim(col("name")).alias("trim_name"),
upper(col("name")).alias("upper_name"),
lower(col("name")).alias("lower_name"),
initcap(col("name")).alias("initcap_name")


)
df_string.show(truncate=False)
```

```python
df_string.show(5)
```

```python
df_string.select('id','name','email',
                 length(col("name")).alias("name_length"),\
                    substring(col("name"), 1, 6).alias("sub_name"),\
                    concat(trim(col("name")), lit(" , "), col("id")).alias("concat_col"),\
                     split(col("email"), "@").getItem(1).alias("email_domain")
                                   
                ).show(truncate=False)

```

# Date functions

```python
data = [
(1, "2025-01-10"),
(2, "2025-02-15"),
(3, "2025-03-20")
]

schema = ['id', 'order_date']

df = spark.createDataFrame(data, schema)
print(df.printSchema()) #it should be string type only

df_date = df.withColumn(
                        "order_date", to_date(col("order_date"), "yyyy-MM-dd")
                    ).select(
                            "id",
                            "order_date",
                            current_date().alias("today"),
                            date_add(col("order_date"), 7).alias("plus_7 _days"),
                            date_sub(col("order_date"), 7).alias("minus_7 _days"),
                            datediff(current_date(), col("order_date")).alias("days_diff"),
                            year(col("order_date")).alias("year"),
                            month(col("order_date")).alias("month"),
                            dayofmonth(col("order_date")).alias("day"),
                            date_format(col("order_date"), "yyyy/MM/dd").alias("formatted_date")
                            )
df_date.show(truncate=False)
```

# union , unionAll and unionByName
Used to append rows of one DataFrame to another.

```python
data1 = [
(1, "PLACED", 100),
(2, "SHIPPED", 200),
(3, "DELIVERED", 300)
]

data2 = [
(1, "PLACED", 100),
(4, "PLACED", 150)
]

schema="id int,status string,price int"

df1= spark.createDataFrame(data1,schema=schema)
df2=spark.createDataFrame(data2,schema=schema)

df1.show()
df2.show()
df1.union(df2).show()

df1.unionAll(df2).show()
```

```python
#wrong placement of columns but datatypes are compatible 

data1 = [
(1, "PLACED", 100),
(2, "SHIPPED", 200),
(3, "DELIVERED", 300)
]

data2 = [
(1, 100,"PLACED"),
(4, 500,"PLACED")
]

schema="id int,status string,price int"

df1= spark.createDataFrame(data1,schema=schema)
df2=spark.createDataFrame(data2,schema='id int,price int,status string')

df1.show()
df2.show()
df1.union(df2).show()

df1.unionAll(df2).show()
# fixed
df2_fixed = df2.select('id','status','price')
df1.union(df2_fixed).show()
```

```python
#less no of columns

data1 = [
(1, "PLACED", 100),
(2, "SHIPPED", 200),
(3, "DELIVERED", 300)
]

data2 = [
(1,"PLACED"),
(4,"SHIPPED")
]

schema="id int,status string,price int"

df1= spark.createDataFrame(data1,schema=schema)
df2=spark.createDataFrame(data2,schema='id int,status string')

df1.show()
df2.show()
df1.union(df2).show()

df1.unionAll(df2).show()
```

# unionByName()

```python

data1 = [
(1, "PLACED", 100),
(2, "SHIPPED", 200),
(3, "DELIVERED", 300)
]

data2 = [
(1, 100,"PLACED"),
(4, 500,"PLACED")
]

schema="id int,status string,price int"

df1= spark.createDataFrame(data1,schema=schema)
df2=spark.createDataFrame(data2,schema='id int,price int,status string')

df1.show()
df2.show()
df1.union(df2).show()

df1.unionByName(df2).show()
```

```python
df1.unionByName(df2).show()

df1.unionByName(df2).distinct().show()

df1.unionByName(df2).dropDuplicates().show()
```

# intersect() ,intersectAll()

```python
data1 = [
            (1, "PLACED", 100),
            (2, "SHIPPED", 200),
            (3, "DELIVERED", 300),
            (3, "DELIVERED", 300),
            (3, "DELIVERED", 300)
        ] 

data2 = [
            (3, "DELIVERED", 300),
            (3, "DELIVERED", 300),
            (4, "PLACED", 150)
        ]

order_schema = "order_id int, order_status string, amount int"
df1 = spark.createDataFrame(data1, order_schema)
df2 = spark.createDataFrame(data2, order_schema)

df1.intersect(df2).show()

df1.intersectAll(df2).show()
```

# exceptAll / subtract

```python
df1.show()
df2.show()

df1.subtract(df2).show() #present in df1 only,removes duplicate
df1.exceptAll(df2).show() #keeps duplicate

df2.subtract(df1).show()
```

# Aggregation with group by

```python
# state wise whats my sales
```

```python
df_orders.show(5)
```

```python
df_orders.groupBy('state').agg(sum('total_price').alias('stateWiseSales')).show(truncate=False)
```

```python
# state wise no of orders
df_orders.groupBy('state')\
            .agg(count('order_id').alias('total_orders'))\
            .sort(col('total_orders').desc())\
            .show()
```

```python
 # statewise order status volume
```

```python
df_orders.groupBy('state','order_status')\
            .agg(count('*').alias('volume'))\
            .orderBy('state','order_status')\
            .show()
```

```python
df_orders.groupBy('state','order_status')\
            .agg(count('*').alias('volume'))\
            .where(col('state')=='Goa')\
            .show()
```

# take

```python
df_orders.groupBy('state','order_status')\
            .agg(count('*').alias('volume'))\
            .where(col('state')=='Goa')\
            .take(3)
```

```python
# top 3 states where orders are max
df_orders.groupBy('state')\
            .agg(count('order_id').alias('total_orders'))\
            .sort(col('total_orders').desc())\
            .take(3)
```

```python
df_orders.groupBy('state')\
            .agg(count('order_id').alias('total_orders'),sum('quantity').alias('total_quantity'))\
            .sort(col('total_orders').desc())\
            .show(25)
```

# where/filter

```python
#states where no of orders is lesser than 45100

df_orders.groupBy('state')\
            .agg(count('order_id').alias('total_orders'))\
            .where(col('total_orders')<45100)\
            .show()
```

```python

df_orders.groupBy('state')\
            .agg(count('order_id').alias('total_orders'))\
            .filter(col('total_orders')<45100)\
            .show()
```
