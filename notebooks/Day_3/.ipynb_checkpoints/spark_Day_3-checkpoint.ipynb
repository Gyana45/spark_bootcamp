## Partition scenario

```python
schema='order_id int,customer_id int,product_id int,price float, order_date date,order_status string,state string,quantity int'

df = spark.read.format('csv')\
                .option('header','true')\
                .schema(schema)\
                .load('/data/orders_300mb.csv')


df.rdd.getNumPartitions() # output--4
#file size is 335mb---128,128,79--3 tasks--but 4 tasks will be created
```

```python
# DRY RUN
df.write.mode('overwrite').format('noop').save() #noop--no operation is done
#1 job will be created,i have 2 executors,each have 2 cores
#in that job ,4 tasks will be created each willl be around 83mb
```

```python
schema='order_id int,customer_id int,product_id int,price float, order_date date,order_status string,state string,quantity int'

df = spark.read.format('csv')\
                .option('header','true')\
                .schema(schema)\
                .load('/data/orders_50mb.csv')
# DRY RUN
df.write.mode('overwrite').format('noop').save()

# 1.after action mode 1 job will be created
#2.here only 1 partitions required(55mb/128)
#3.we have 2 executors and 2 cores each.so 4 partitions
#4.so 4 tasks/partitions will be created each has around 10mb data
```

```python


schema='order_id int,customer_id int,product_id int,price float, order_date date,order_status string,state string,quantity int'

df = spark.read.format('csv')\
                .option('header','true')\
                .schema(schema)\
                .load('/data/orders_1gb.csv')

df.rdd.getNumPartitions() #9
# df.show() # 1 job--1 stage--1 partition

# # DRY RUN
# df.write.mode('overwrite').format('noop').save() # 1 job--1 stage--9 partitions

df = df.filter("state =='Delhi' ")
df.show() #1job--1stage--1 partition



```

```python
schema='order_id int,customer_id int,product_id int,price float, order_date date,order_status string,state string,quantity int'

df = spark.read.format('csv')\
                .option('header','true')\
                .schema(schema)\
                .load('/data/orders_1gb.csv')

df = df.groupby(['state']).count() #wide transformation,1 extra stage,200 partitions
df.write.mode('overwrite').format('noop').save()  # 1job--1stage(shuffle from one executor to other)--9 partitions
# after shuffle 200 partitions will be created by default
#total 2 stages 
```
